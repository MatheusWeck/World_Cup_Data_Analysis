```python
#Limpando dados NaN da tabela jogos, onde a coluna 'Year" é NaN, e ignorando os 2 valores NaN na coluna 'Attendence'
jogos.dropna(subset=['Year'],inplace=True)
jogos
```

```python
#Procura valores diferentes e unicos em cada coluna(outliers)
for coluna in jogos:
    count_jogos = jogos[coluna].value_counts()
    print(count_jogos)
 
# Encontrado nomes errados na coluna Home Team Name
# Encontrado nomes errados na coluna Stadium
```

```python
#Corrigindo textos
jogos_corrigidos = jogos.replace('rn">', '', regex=True)
jogos_corrigidos = jogos_corrigidos.replace('Nou Camp - Estadio Leï¿½n', 'Nou Camp - Estadio León')
jogos_corrigidos = jogos_corrigidos.replace('Stade Vï¿½lodrome', 'Stade Vélodrome')
jogos_corrigidos = jogos_corrigidos.replace('Estadio Olï¿½mpico Chateau Carreras', 'Estadio Olímpico Chateau Carreras')
jogos_corrigidos = jogos_corrigidos.replace('Maracanï¿½ - Estï¿½dio Jornalista Mï¿½rio Filho', 'Maracanã - Estádio Jornalista Mário Filho')
jogos_corrigidos = jogos_corrigidos.replace('Estadio Josï¿½ Marï¿½a Minella', 'Estadio José María Minella ')
jogos_corrigidos = jogos_corrigidos['Year'].astype(int)
jogos_corrigidos

#jogos_corrigidos['Home Team Name'].value_counts()
#jogos_corrigidos['Stadium'].value_counts().head(50)
```
